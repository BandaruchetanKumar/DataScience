```python
import json
df3['genres'] = df3['genres'].apply(lambda x: json.loads(x))
# Create a new column 'genre_name' to extract genre names
df3['genre_name'] = df3['genres'].apply(lambda x: [genre['name'] for genre in x])
# Explode the 'genre_name' column to create a row for each genre
exploded_data = df3.explode('genre_name')
# Group by genre and calculate the average vote for each genre
average_vote_per_genre = exploded_data.groupby('genre_name')['vote_average'].mean().sort_values(ascending=False)
# Plotting the graph
plt.figure(figsize=(12, 6))
average_vote_per_genre.plot(kind='bar', color='skyblue')
plt.title('Average Vote by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Vote')
plt.show()
```
